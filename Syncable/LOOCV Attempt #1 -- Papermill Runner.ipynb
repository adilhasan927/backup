```python
import papermill as pm
from IPython.display import display, clear_output
```

```python
bander1 = "lambda x: x[:,4]"
```

```python
pm.execute_notebook(
   r'LOOCV Attempt #1 -- Papermilled.ipynb',
   r'LOOCV Attempt 1 only means filter.ipynb',
   parameters=dict(bander = bander1)
)
clear_output(wait=False)
```

```python
bander2 = "lambda x: x[tuple(zip(*[(2,0), (2,4), (3,4), (4,1), (5,0), (5,4), (6,1), (7,4)]))]"
```

```python
pm.execute_notebook(
   r'LOOCV Attempt #1 -- Papermilled.ipynb',
   r'LOOCV Attempt 1 selected filters.ipynb',
   parameters=dict(bander = bander2)
)
clear_output(wait=False)
```

```python
bander2_minus_means = "lambda x: x[tuple(zip(*[(2,0), (4,1), (5,0), (6,1)]))]"
bander3 = "lambda x: np.concatenate(({}, {}))".format(
    bander1[10:], bander2_minus_means[10:])
pm.execute_notebook(
   r'LOOCV Attempt #1 -- Papermilled.ipynb',
   r'LOOCV Attempt 1 selected + means filter.ipynb',
   parameters=dict(bander = bander3)
)
clear_output(wait=False)
```

```python
bander4 = "lambda x: x.flatten()"
pm.execute_notebook(
   r'LOOCV Attempt #1 -- Papermilled.ipynb',
   r'LOOCV Attempt 1 all filters.ipynb',
   parameters=dict(bander = bander4)
)
clear_output(wait=False)
```
